### ticks,labels,legends

```python
fig=plt.figure(figsize=(10,8))
ax=fig.subplots(1,1)
ax.plot(np.random.randn(1000).cumsum())
ax.set_xticklabels(["one","two","three","four","five","six","seven"],
                  rotation=60)
ax.set_title("line chart")
ax.set_xlabel("year")
```

```python
fig,ax=plt.subplots(2,2,sharex=True,sharey=True)
ax[0, 0].plot(np.random.randn(20),np.random.randn(20))
ax[0,1].hist(np.random.randn(20).cumsum(),color="red")
ax[1,0].scatter(np.random.randn(20),np.random.randn(20))
ax[1,1].plot(np.random.randn(20).cumsum(),linestyle="--",marker="o",color="black")
plt.show()
```

```python
#also we can set a properties 
fig=plt.figure(figsize=(10,8))
props={
    "title":"represent line plot",
    "xlabel":"year",
    "ylabel":"profit"
}
ax=fig.subplots(1,1)
ax.plot(np.random.randn(1000).cumsum())
ax.set(**props)
```

```python
#adding the legend
fig=plt.figure(figsize=(10,7))
ax=fig.subplots(1,1)
ax.plot(np.random.randn(1000).cumsum(),color="red",linestyle="--",label="one")
ax.plot(np.random.randn(1000).cumsum(),color="blue",linestyle="solid",label="two")
ax.plot(np.random.randn(1000).cumsum(),color="black",linestyle="dotted",label="three")
ax.legend(loc="best")           #The loc tells matplotlib where to place the plot
```
