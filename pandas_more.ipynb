```python
data=pd.DataFrame({"d1":np.random.randn(1000),
                  "d2":np.random.randn(1000)})
```

```python
quat=pd.cut(data.d1,4)
quat
```

```python
def get_stat(group):
    return {"mean":np.mean(group),
           "std":np.std(group),
           "min":np.min(group),
           "max":np.max(group)}
```

```python
grouped=data.d1.groupby(quat)
```

```python
grouped.apply(get_stat).unstack()
```

```python
q_cut=pd.qcut(data.d1,5)
q_cut
```

```python
data.d1.groupby(q_cut).apply(get_stat).unstack()
```

```python
df=pd.DataFrame({"country":["ind","pak","ind","uk","uk","ind","pak","usa"],
                "d1":np.random.randn(8)})
df.iloc[[2,3,6],1]=np.nan
df
```

```python
suits=["H","S","D","F"]
value=(list(range(2,11))+["A","J","K","Q"])*4
base_names =list(range(2, 11)) + ['A','J', 'K', 'Q']
print(base_names)
card=[]
for i in suits:
        card.extend(str(num)+i for num in base_names)
```

```python
deck=pd.Series(value,index=card)
deck[:13]
```

```python
def draw(deck,n=5):
    return deck.sample(n)
```

```python
draw(deck,n=4)
```

```python
get_suit=lambda x:x[-1]    #return the last character from the given series
```

#### want to draw a n cards from each suits

```python
#return two cards from each suits.
deck.groupby(get_suit).apply(draw,n=2)
```

### groupby weighted average  and correlations

```python
df_weighted=pd.DataFrame({"code":["a","a","b","a","b","b","a","b"],
                         "data":np.random.randn(8),
                         "weights":np.random.rand(8)})
df_weighted
```

```python
get_weighted_avg=lambda x:np.average(x["data"],weights=x["weights"])
```

```python
df_weighted.groupby("code").apply(get_weighted_avg)
```

```python
df_stock=pd.read_csv("C:/nidhisha/dataset/stock_px_2.csv",)
df_stock.head(3)
```

```python
date=pd.to_datetime(df_stock["Unnamed: 0"])
df_stock.drop(labels="Unnamed: 0",axis=1,inplace=True)
df_stock.index=date
```

```python
df_stock.info()
```

```python
get_corr=lambda x:x.corrwith(x["SPX"])
```

#### pct_change()-  function is commonly used to calculate the percentage change between the current and a prior element in a DataFrame or a Series.

```python
rets=df_stock.pct_change().dropna()
```

```python
get_year=lambda x:x.year  # here we are getting date type of data extracting oly the year
```

```python
rets.groupby(get_year).apply(get_corr)     # here group based on year wise and caculating the correlation by year with SPX col
```
